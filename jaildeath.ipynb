```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.patches import Patch
```

```python
df = pd.read_csv("all_deaths.csv", encoding='latin-1')
```

```python
df_2008 = df[df["year"] == 2008].sort_values("date_of_death")
df_2009 = df[df["year"] == 2009].sort_values("date_of_death")
df_2010 = df[df["year"] == 2010].sort_values("date_of_death")
df_2011 = df[df["year"] == 2011].sort_values("date_of_death")
df_2012 = df[df["year"] == 2012].sort_values("date_of_death")
df_2013 = df[df["year"] == 2013].sort_values("date_of_death")
df_2014 = df[df["year"] == 2014].sort_values("date_of_death")
df_2015 = df[df["year"] == 2015].sort_values("date_of_death")
df_2016 = df[df["year"] == 2016].sort_values("date_of_death")
df_2017 = df[df["year"] == 2017].sort_values("date_of_death")
df_2018 = df[df["year"] == 2018].sort_values("date_of_death")
df_2019 = df[df["year"] == 2019].sort_values("date_of_death")
```

```python
all_deaths = [len(df_2008),len(df_2009),len(df_2010),len(df_2011),len(df_2012),len(df_2013),len(df_2014),len(df_2015),len(df_2016),len(df_2017),
                     len(df_2018),len(df_2019)]
all_deaths
```

```python
percent_diff_year2year = []
for count in range (len(all_deaths) - 1):
    percent = 0
    percent = round((all_deaths[count + 1] - all_deaths[count]) / ((all_deaths[count + 1] + all_deaths[count]) /  2), 3)
    percent_diff_year2year.append(percent * 100)
```

```python
percent_diff_year2year
```

```python
percent_diff_from_2008 = []
for count in range (len(all_deaths) - 1):
    percent = 0
    percent = round((all_deaths[1 + count] - all_deaths[0]) / ((all_deaths[1 + count] + all_deaths[0]) /  2), 3)
    percent_diff_from_2008.append(percent * 100)
```

```python
percent_diff_from_2008
```

```python
labels = ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
sns.set(style='darkgrid')
fig, ax = plt.subplots()
sns.barplot(x=labels, y=percent_diff_from_2008, ax=ax)
ax.set_ylabel('Percentage')
ax.set_title('Percentage Increase in Jail Deaths From 2008')
plt.show()
```

```python
labels = ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
sns.set(style='darkgrid')
fig, ax = plt.subplots()
sns.barplot(x=labels, y=percent_diff_year2year, ax=ax)
ax.set_ylabel('Percentage')
ax.set_title('Percentage Change in Jail Deaths From Prior Year')
plt.show()
```

```python
mf_df = df[df['gender'].isin(['M', 'F'])]
```
